```python
np.random.seed(0)
data = np.random.rand(100, 2)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

X = scaled_data[:, :]
wcss = []

for n in range(1, 11):
    algorithm = KMeans(n_clusters=n, init='k-means++', n_init=10, max_iter=300, random_state=0)
    algorithm.fit(X)
    wcss.append(algorithm.inertia_)

plt.figure(1, figsize=(15, 6))
plt.plot(np.arange(1, 11), wcss, 'o')
plt.plot(np.arange(1, 11), wcss, '-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method Diagram')

optimal_n = 3
kmeans = KMeans(n_clusters=optimal_n, init='k-means++', n_init=10, max_iter=300, random_state=0)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

plt.legend()
plt.show()
```

```python
plt.figure(2, figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X')
plt.title(f'Cluster Visualization with 3 Clusters')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,10))

plt.subplot(2,2,1)

population.boxplot()

plt.xticks(rotation = 70)
# Customize the plot (optional)
plt.title("Boxplot   ")
plt.xlabel("Columns")
plt.ylabel("Values")

plt.subplot(2,2,2)
#plt.plot()
filtered_data = population[(population['Year'] >= 1995) & (population['Year'] <= 2020)]

# Select columns for clustering: Year and Population
year_population = filtered_data[['Year', 'Population']]

# Scale the data
scaler = StandardScaler()
year_population_scaled = scaler.fit_transform(year_population)

# Perform KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
filtered_data['Cluster'] = kmeans.fit_predict(year_population_scaled)

for cluster in range(3):
    cluster_data = filtered_data[filtered_data['Cluster'] == cluster]
    plt.scatter(cluster_data['Year'], cluster_data['Population'], label=f'Cluster {cluster}')

plt.title('Clustering by Year and Population (1995-2020)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid()
#plt.tight_layout()

plt.subplot(2,2,3)
plt.plot()

plt.subplot(2,2,4)
plt.plot()

plt.show()
```
